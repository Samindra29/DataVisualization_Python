## Pie chart for hypothetical parts of company shares

```python
# defining a list for pie portions
values = [20,55,5,17,3]

# assigning colors
colors = ['g', 'r', 'b', 'y', 'm']

# lables to be assigned to the pies
labels = ['APPLE', 'GOOGLE', 'TESLA', 'AMAZON', 'NETFLIX']

# if we want to highlight or explode 
# define the separation in the postions
explode = [0, 0.1, 0, 0,0.1]

# plotting
plt.figure(figsize=(10,15))
plt.title('STOCK PORTFOLIO')

# explode here is optional parameter
plt.pie(values, colors = colors, labels=labels, explode=explode)

plt.show()
```

## Histogram for daily sales

```python
# bins = no. of bars we want to plot. default is 10
num_bins = 40

# if we want to show and calculate miu and sigma
mu = daily_return_df['FB'].mean()
sigma = daily_return_df['FB'].std()

plt.figure(figsize=(10,10))
plt.grid()
plt.hist(daily_return_df['FB'], num_bins, facecolor='blue');

plt.title(f'Histogram: mu = {mu} , sigma = {sigma}');
```

# Visualization with Seaborn

```python
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
cancer
```

```python
# created dataset (pre-processed)
df_cancer = pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns = np.append(cancer['feature_names'], ['target']))
```

```python
df_cancer
```

```python
df_cancer.head(10)
```

```python
# last 10 rows of the dataset
df_cancer.tail(10)
```

```python
plt.figure(figsize = (10,10))

# plot scatter plot for cancer area and tumour smoothness against target(malignant or benign)
# hue is the colour code to seprate the targets
sns.scatterplot(x = 'mean area', y = 'mean smoothness', hue = 'target', data = df_cancer)
```

```python
plt.figure(figsize = (10,10))

sns.scatterplot(x = 'mean radius', y = 'mean area', hue = 'target', data = df_cancer)
```

## Pairplot

```python
sns.pairplot(df_cancer, hue = 'target', vars = ['mean radius', 'mean texture', 'mean area', 'mean perimeter', 'mean smoothness'])
```

## plotting correlation

```python
plt.figure(figsize = (30,30))
sns.heatmap(df_cancer.corr(), annot = True);
```

```python
# Displot combines matplotlib histogram function with kdeplot() (Kernel density estimate)
# KDE is used to plot the Probability Density of a continuous variable. 

sns.distplot(df_cancer['mean radius'], bins = 25, color = 'b');
```

```python
# plotting 2 distplot for the 2 targets

# need to separate the 2 df where target = 1 and where target = 0
class_0_df = df_cancer[ df_cancer['target'] == 0]
class_1_df = df_cancer[ df_cancer['target'] == 1]
```

```python
sns.distplot(class_0_df['mean radius'], bins = 25, color = 'b');
sns.distplot(class_1_df['mean radius'], bins = 25, color = 'r');
plt.grid()
```
